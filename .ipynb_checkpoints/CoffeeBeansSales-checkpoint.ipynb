# Introduction
The dataset involves a Coffee beans Sales Raw xlsx file with three sheets:
- customers
- orders 
- products

## Problem Statement
- To uncover trends in coffee consumption by analysing order patterns over time
- Create customer segments based on demographics and preferences.
- Identify the most profitable coffee products and optimize pricing.
- Ensure that the right quantities of each coffee product are stocked.
- Tailor marketing campaigns to specific customer segments.

# Dataset 

- Orders Worksheet:

    - Order ID: A unique identifier for each coffee order.
    - Order Date: The date when the order was placed.
    - Customer ID: An identifier linking the order to a specific customer.
    - Product ID: A unique identifier for each coffee product.
    - Quantity: The quantity of the coffee product ordered.

- Customers Worksheet:

    - Customer ID: A unique identifier for each customer.
    - Customer Name: The name of the customer.
    - Email Address: Contact information for customers.
    - Phone Number: Another contact detail for customers.

- Products Worksheet:

    - Product ID: A unique identifier for each coffee product.
    - Coffee Type: The type or blend of coffee, such as Arabica or Robusta.
    - Roast Type: The roast level, including light, medium, or dark roast.
    - Size: Information about the product size.
    - Unit Price: The price of a single unit of the coffee product.
    - Price Per 100g: The price per 100 grams for detailed price comparisons.
    - Profit: Insights into the profitability of each coffee product.

# Data Cleaning
- This method will check the missing values, if the respective coumns are not important they can be completely avoided while if not the missing data could be dropped.
- The redundant columns within the orders data needs to dealt with.
- check and fix the datatypes to appropriate formats // Since we took a xlsx file.

```python
# Dropping rows with missing 'Email' or 'Phone Number' if these fields are necessary

df_customers_cleaned = df_customers.dropna(subset=['Email', 'Phone Number'])

# Check the cleaned dataframe
df_customers_cleaned
```

```python
# drop the irrelevant columns in the orders datasheet 
df_orders_cleaned = df_orders.drop(columns=['Customer Name', 'Email', 'Country', 'Coffee Type', 'Roast Type', 'Size', 'Unit Price', 'Sales'])

#check the cleaned dataframe
df_orders_cleaned
```

```python
# Convert 'Order Date' to datetime
df_orders_cleaned['Order Date'] = pd.to_datetime(df_orders_cleaned['Order Date'], format= '%d/%m/%Y')

#Verify data types
print('Orders Data (Datatypes): ')
df_orders_cleaned.dtypes
```

# Data analysis
- For proper analysis of the datasets, we can combine the each of the dataframes: 'orders', 'customers', 'prodcuts'
- Perform some initial exploratory analysis to understand the data better, such as calculating sales, analyzing customer behavior, and understanding product performance.

```python
# Merge 'orders' and 'customers' on 'Customer ID'
df_orders_customers = pd.merge(df_orders_cleaned, df_customers_cleaned, on='Customer ID', how='left')

# Merge the resulting merged dataframe with  'products' on 'Product ID'
df_complete = pd.merge(df_orders_customers, df_products, on='Product ID', how='left')

#check the final dataframe
df_complete.head()
```

## Data Visualization
- Sales over Time
- Top Products 
- Customer Segmentation


```python
print('Sales Over time:')

# Create a new columns calculating the sales for the sales over time graph 
df_complete['Sales'] = df_complete['Quantity'] * df_complete['Unit Price']

# For daily sales the grouping by order date 
sales_over_time = df_complete.groupby('Order Date')['Sales'].sum().reset_index()

#Plot the graph
plt.figure(figsize=(12,6)) #size of the graph
plt.plot(sales_over_time['Order Date'], sales_over_time['Sales'], marker='o') # enter data for ploting
plt.title('Sales Over Time') # add headr of the graph
plt.xlabel('Order Date')# add label for the x-axis values
plt.ylabel('Sales')# add label for y-axis values
plt.grid(True)# add graph into the chart to visualize the data values easier
plt.show() # display the dataset


```

```python
print('Top Sellin products: ')

# find the top ten selling products
top_products = df_complete.groupby('Product ID')['Sales'].sum().sort_values(ascending=False).head(10)

#plot the graph
plt.figure(figsize=(10,6)) #size of the graph
top_products.plot(kind='bar', color='blue')# enter data for ploting
plt.title('Top 10 Products by sales') # add headr of the graph
plt.xlabel('Prodcut ID')# add label for the x-axis values
plt.ylabel('Total Sales')# add label for y-axis values
plt.grid(True)# add graph into the chart to visualize the data values easier
plt.xticks(rotation=25) # slant headers to stop confusion on differentiating the product Id
plt.show()# display the dataset
```
